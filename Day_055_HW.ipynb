```python
plt.figure(figsize=(5, 5))

y_pred = KMeans(init='random', n_init=1, n_clusters=3, random_state=0).fit_predict(X)

plt.scatter(X[y_pred==0, 0], X[y_pred==0, 1], c='red', label='iris-setosa')
plt.scatter(X[y_pred==1, 0], X[y_pred==1, 1], c='green', label='iris-versicolor')
plt.scatter(X[y_pred==2, 0], X[y_pred==2, 1], c='blue', label='iris-virginia')
plt.title("random init, run algorithm only once")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5, 5))

y_pred = KMeans(init='k-means++', n_clusters=8,  random_state=0).fit_predict(X)

plt.scatter(X[y_pred==0, 0], X[y_pred==0, 1], c='red', label='iris-setosa')
plt.scatter(X[y_pred==1, 0], X[y_pred==1, 1], c='green', label='iris-versicolor')
plt.scatter(X[y_pred==2, 0], X[y_pred==2, 1], c='blue', label='iris-virginia')
plt.title("8 clusters")

plt.legend()
plt.show()
```
